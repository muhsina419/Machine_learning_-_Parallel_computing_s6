```python
import pandas as pd
import numpy as np
from datetime import datetime, timezone
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
```

```python
# Suppress DeprecationWarning from jupyter_client
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Record analysis start time
current_time = datetime.now(timezone.utc)
print(f"Analysis started at: {current_time}\n")
```

```python
# Dataset URL
url = "https://raw.githubusercontent.com/earlyann/grocery_market_basket_analysis/master/Groceries_dataset.csv"

# Load dataset
df_raw = pd.read_csv(url)
```

```python
print("Dataset Structure:")
print(df_raw.info())

print("\nFirst 5 rows of transactional data:")
print(df_raw.head())
```

```python
# Group by Member_number and Date
# Each group represents one transaction
transactions = df_raw.groupby(['Member_number', 'Date'])['itemDescription'] \
                     .apply(list) \
                     .values.tolist()
```

```python
# Convert transactions into binary format
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print(f"\nTotal Transactions: {len(transactions)}")
print(f"Total Unique Items: {len(te.columns_)}")
```

```python
# Set minimum support threshold
min_support_threshold = 0.001

# Apply Apriori algorithm
frequent_itemsets = apriori(
    df_encoded,
    min_support=min_support_threshold,
    use_colnames=True
)

print("\nTop 5 Frequent Itemsets:")
print(frequent_itemsets.sort_values(by='support', ascending=False).head())
```

```python
# Set minimum confidence threshold
min_confidence = 0.05

# Generate rules
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=min_confidence
)
```

```python
# Filter rules with Lift > 1
significant_rules = rules[rules['lift'] > 1.0] \
                    .sort_values(by='lift', ascending=False)

print("\nFinal Association Rules Evaluation")

# Display top 10 rules
print(significant_rules[['antecedents',
                         'consequents',
                         'support',
                         'confidence',
                         'lift']].head(10))
```
